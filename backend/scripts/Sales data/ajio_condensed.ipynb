```python
def format_data(df, colnames, skip_last=0):
    
    # if 'Agreed Margin %' in colnames:
    #     df['margin'] = (df['Agreed Margin %'] * df['Net sales with tax +Own fund'])
    # else:

    df = df[colnames]
    
    df['date'] = pd.to_datetime(df[colnames[0]])
    df['ean'] = df[colnames[1]]
    # todo: mrp = gross sales / no of items
    df['mrp'] = df[colnames[2]].astype('float')
    df['qty'] = df[colnames[3]]
    df['sp'] = df[colnames[4]].astype('float')
    df['margin'] = df[colnames[5]].astype('float')
    # print
    df = df[['date', 'ean','qty','mrp','sp', 'margin']]
    if skip_last:
        return df[:-skip_last]
    return df
```

```python
condensed_df = pd.DataFrame(columns=['date', 'ean', 'sold_qty', 'return_qty', 'net_qty', 'avg_mrp', 'avg_discount', 'avg_sp', 'avg_marketplace_margin'])

for k, v in condensed_data.items():
    sold_qty = 0
    return_qty = 0
    avg_mrp = 0
    avg_discount = 0
    avg_sp = 0
    avg_marketplace_margin = 0
    no_of_items = 0

    for i in v:
        no_of_items += 1
        if i[0] >= 0:
            sold_qty += abs(i[0])
        else:
            return_qty += abs(i[0])
        avg_mrp += abs(i[1])
        avg_sp += abs(i[2])
        avg_marketplace_margin += abs(i[3])

    if not avg_mrp:
        continue
    avg_mrp /= no_of_items
    avg_sp /= no_of_items
    avg_marketplace_margin /= no_of_items
    net_qty = sold_qty - return_qty   
    avg_discount = (avg_mrp - avg_sp) * 100 / avg_mrp 
    condensed_df = condensed_df.append({'date': k.split('/')[0], 'ean': k.split('/')[1], 'sold_qty': sold_qty, 'return_qty': return_qty, 'net_qty': net_qty, 'avg_mrp': avg_mrp, 'avg_discount': avg_discount, 'avg_sp': avg_sp, 'avg_marketplace_margin': avg_marketplace_margin}, ignore_index=True) # type: ignore
    # break
# condensed_df
```
