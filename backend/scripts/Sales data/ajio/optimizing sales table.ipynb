```python
import pandas as pd
```

```python
with open('../../../data/ajio_sales_with_product_id.csv') as f:
    df = pd.read_csv(f)
```

```python
df
```

```python
sales = df[df['status']==1]
returns = df[df['status']==-1]
```

```python
salesmean = sales.groupby(['sale_date','product_id']).mean()
salesmean.drop('ean', axis=1, inplace=True)
returnsmean = returns.groupby(['sale_date','product_id']).mean()
returnsmean.drop('ean', axis=1, inplace=True)
returnsmean
```

```python
salesmean.reset_index(inplace=True)
returnsmean.reset_index(inplace=True)
df = pd.concat([salesmean, returnsmean], ignore_index=True)
df.sort_values(['sale_date','product_id'], inplace=True)
df
```
