```python

# Load the CSV files into pandas dataframes
soh_df = pd.read_csv('../static/cleaned_updated_soh.csv')
sales_df = pd.read_csv('../static/cleaned_updated_sales.csv')

# Convert the date columns to datetime objects
soh_df['date'] = pd.to_datetime(soh_df['date'], format='%d-%m-%Y', errors='coerce')
sales_df['date'] = pd.to_datetime(sales_df['date'], format='%d-%m-%Y', errors='coerce')

# Define the chosen month (for example, April 2023)
chosen_month = pd.Timestamp('2023-12-01')

# Step 1: Calculate total stock for each style_id for the chosen month
soh_monthly = soh_df[soh_df['date'].dt.to_period('M') == chosen_month].groupby('style_id')['qty'].sum().reset_index()
soh_monthly_sorted = soh_monthly.sort_values(by='qty', ascending=False)


# Define the first day of the three months preceding the chosen month
three_months_prior = [(chosen_month - pd.DateOffset(months=i)).replace(day=1) for i in range(1, 4)]

# Filter the sales data for the three months preceding the chosen month
filtered_sales = sales_df[sales_df['date'].dt.to_period('M').isin([m.to_period('M') for m in three_months_prior])]

# Aggregate sales data by style_id and month
monthly_sales = filtered_sales.groupby(['style_id', pd.Grouper(key='date', freq='M')])['qty'].sum().reset_index()

# Convert date to month name
monthly_sales['month'] = monthly_sales['date'].dt.strftime('%B')

# Pivot the data to have each month's sales in separate columns
pivoted_sales = monthly_sales.pivot(index='style_id', columns='month', values='qty').fillna(0)

# Add a new column for the average sales of these 3 months
pivoted_sales['avg_sales_3m'] = pivoted_sales.mean(axis=1)

# Sort the dataframe by avg_sales_3m in descending order
sorted_pivoted_sales = pivoted_sales.sort_values('avg_sales_3m', ascending=False)

# Display the sorted pivoted sales data with the average
print(sorted_pivoted_sales)
```

```python
# Load the CSV files into pandas dataframes
soh_df = pd.read_csv('../static/cleaned_updated_soh.csv')
sales_df = pd.read_csv('../static/cleaned_updated_sales.csv')

# Convert the date column to a datetime object
sales_df['date'] = pd.to_datetime(sales_df['date'], format='%d-%m-%Y', errors='coerce')
soh_df['date'] = pd.to_datetime(soh_df['date'], format='%d-%m-%Y', errors='coerce')

# Define the chosen month and previous three months
chosen_month = pd.Timestamp('2023-11-01')
three_months_prior = [(chosen_month - pd.DateOffset(months=i)).replace(day=1) for i in range(1, 4)]

# Filter the sales data for the three months preceding the chosen month
filtered_sales = sales_df[sales_df['date'].dt.to_period('M').isin([m.to_period('M') for m in three_months_prior])]

# Aggregate sales data by style_id and month
monthly_sales = filtered_sales.groupby(['style_id', pd.Grouper(key='date', freq='M')])['qty'].sum().reset_index()
monthly_sales['month'] = monthly_sales['date'].dt.strftime('%B')
pivoted_sales = monthly_sales.pivot(index='style_id', columns='month', values='qty').fillna(0)
pivoted_sales['avg_sales_3m'] = pivoted_sales.mean(axis=1)
sorted_pivoted_sales = pivoted_sales.sort_values('avg_sales_3m', ascending=False)

# Filter and aggregate the SOH data for the chosen month
filtered_soh = soh_df[soh_df['date'].dt.to_period('M') == chosen_month.to_period('M')]
total_stock = filtered_soh.groupby('style_id')['qty'].sum().reset_index()

# Merge the SOH data with the sorted sales data on style_id
combined_data = pd.merge(sorted_pivoted_sales, total_stock, on='style_id', how='left')

# Add the new column soh/avg_sales_3m and perform the division where values are not NaN
# combined_data['soh/avg_sales_3m'] = combined_data.apply(lambda x: x['qty'] / x['avg_sales_3m'] if pd.notna(x['qty']) and pd.notna(x['avg_sales_3m']) else None, axis=1)

# Add the new column soh/avg_sales_3m and perform the division where values are not NaN
combined_data['soh/avg_sales_3m'] = combined_data.apply(
    lambda x: x['qty'] / x['avg_sales_3m'] if pd.notna(x['qty']) and pd.notna(x['avg_sales_3m']) else None, axis=1
)

# Add columns for overstocked and understocked based on 'soh/avg_sales_3m'
combined_data['overstocked'] = combined_data['soh/avg_sales_3m'] >= 6
combined_data['understocked'] = combined_data['soh/avg_sales_3m'] <= 2

# Display the combined data
print(combined_data)

# Save the combined data to a CSV file
combined_data.to_csv('../static/stock_analysis.csv', index=False)
```
