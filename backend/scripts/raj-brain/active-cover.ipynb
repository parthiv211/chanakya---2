```python
import pandas as pd
pd.set_option('display.max_rows', 200)
master = pd.read_csv('../../data/rajbrain/sud/ean-product-map.csv')
master['option'] = master['Style No'] + '-' + master['Prominent Colour']
master
```

# Total SOH

```python
# calculate Total SOH for this month (23-06) at ean level
df1 = pd.read_csv('../../data/rajbrain/sud/SOH/ck-palya/23-06.csv', index_col=0)
df2 = pd.read_csv('../../data/rajbrain/sud/SOH/FF/23-06.csv', index_col=0) # some are negatives
df3 = pd.read_csv('../../data/rajbrain/sud/SOH/Ajio-SOR/23-06.csv', index_col=0)
df4 = pd.read_csv('../../data/rajbrain/sud/SOH/FK-SOR/23-06.csv', index_col=0)
df5 = pd.read_csv('../../data/rajbrain/sud/Intransit/23-06.csv', index_col=0)

SOH_df = df1.add(df2, fill_value=0).add(df3, fill_value=0).add(df4, fill_value=0).add(df5, fill_value=0)
SOH_df = SOH_df.reset_index()
SOH_df = master.merge(SOH_df, on='EAN', how='left')
SOH_df['qty'].fillna(0, inplace=True)
SOH_df.rename(columns={'qty':'Current Month FE Stock'}, inplace=True)
SOH_df.set_index('EAN', inplace=True)
SOH_df.sort_values(by='Current Month FE Stock', ascending=False)
```

```python
# Bestseller SOH for this month at brand-product level
SOH_df[SOH_df['Bestseller']==True][['Brand','Product','Current Month FE Stock']].groupby(['Brand','Product']).sum().sort_values(by='Current Month FE Stock', ascending=False)
```

```python
# Regular SOH for this month at brand-product level
SOH_df[SOH_df['Bestseller']==False][['Brand','Product','Current Month FE Stock']].groupby(['Brand','Product']).sum().sort_values(by='Current Month FE Stock', ascending=False)
```

# 3m Avg sales

```python
# Calculate on avg (3m) how much is sold every month for each ean
avg_sales_df = pd.DataFrame()

for month in ['23-05', '23-04', '23-03']:
    df1 = pd.read_csv(f'../../data/rajbrain/sud/Sales/FF/{month}.csv', index_col=0)
    df2 = pd.read_csv(f'../../data/rajbrain/sud/Sales/FK-SOR/{month}.csv', index_col=0)
    df3 = pd.read_csv(f'../../data/rajbrain/sud/Sales/Ajio-SOR/{month}.csv', index_col=0)
    df4 = pd.read_csv(f'../../data/rajbrain/sud/Sales/MP/{month}.csv', index_col=0)

    avg_sales_df = avg_sales_df.add(df2, fill_value=0).add(df3, fill_value=0).add(df4, fill_value=0).add(df1, fill_value=0)

avg_sales_df = avg_sales_df/3
avg_sales_df.sort_values(by='qty', ascending=False)
```

```python
# Calculate instock% for each option
style_size_map = master.groupby(['option'])['Standard Size'].apply(lambda x: list(set(x))).reset_index()
style_size_map['size_count'] = style_size_map['Standard Size'].apply(lambda x: len(x))
style_size_map.set_index('option', inplace=True)

non_zero_SOH = SOH_df[SOH_df['Current Month FE Stock']!=0]
available_sizes = non_zero_SOH.groupby(['option'])['Standard Size'].apply(lambda x: list(set(x))).reset_index()
available_sizes['size_count'] = available_sizes['Standard Size'].apply(lambda x: len(x))
available_sizes.set_index('option', inplace=True)

instock_df = style_size_map.merge(available_sizes, on='option', how='left').fillna(0)
instock_df['instock%'] = instock_df['size_count_y']*100/instock_df['size_count_x']

instock_df = instock_df.merge(master[['option','Product','Brand','Bestseller']].drop_duplicates(), on='option', how='left')
instock_df
```

```python
# Bestseller Instock% for this month at brand-product level 
instock_df[instock_df['Bestseller']==True][['Brand','Product', 'instock%']].groupby(['Brand','Product']).mean().sort_values(by='instock%', ascending=False)
```

```python
# Regular Instock% for this month at brand-product level 
instock_df[instock_df['Bestseller']==False][['Brand','Product', 'instock%']].groupby(['Brand','Product']).mean().sort_values(by='instock%', ascending=False)
```

# Active Cover

```python
# Calculate active cover for each EAN
active_cover = (SOH_df['Current Month FE Stock']/avg_sales_df['qty']).sort_values(ascending=False).fillna(0)
active_cover = active_cover.reset_index().merge(master[['EAN','Product','Brand','Bestseller']].drop_duplicates(), on='EAN', how='left')
active_cover.rename(columns={0:'active_cover'}, inplace=True)
active_cover
```

```python
# Bestseller Active Cover for this month at brand-product level 
active_cover[active_cover['Bestseller']==True][['Brand','Product', 'active_cover']].groupby(['Brand','Product']).mean().sort_values(by='active_cover', ascending=False)
```

```python
# Regular Active Cover for this month at brand-product level 
active_cover[active_cover['Bestseller']==False][['Brand','Product', 'active_cover']].groupby(['Brand','Product']).mean().sort_values(by='active_cover', ascending=False)
```

```python
# Total Active Cover for this month at brand-product level 
active_cover[['Brand','Product', 'active_cover']].groupby(['Brand','Product']).mean().sort_values(by='active_cover', ascending=False)
```
