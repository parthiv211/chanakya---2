```python
import pandas as pd
pd.options.display.max_columns = None
```

```python
master = pd.read_csv('../../data/backup-styles.csv')
master_imgs = pd.read_csv('../../data/master_with_images.csv')
```

```python
imgs = {}

for index, row in master_imgs.iterrows():
    style_id = row['Style ID']
    url = row['Image URL']
    if not imgs.get(style_id):
        imgs[style_id] = [url]
    else:
        imgs[style_id].append(url)
```

```python
for key in imgs.keys():
    imgs[key] = list(set(imgs[key]))
```

```python
clean_imgs = {}
for key in imgs.keys():
    imgs[key] = [x for x in imgs[key] if type(x) == str]
    if imgs[key]:
        clean_imgs[key] = imgs[key]
```

```python
clean_imgs
```

```python
for index, row in master.iterrows():
    if row['style_id'] in clean_imgs.keys():
        print(row['style_id'])
        master.at[index,'images'] = clean_imgs[row['style_id']]
```
