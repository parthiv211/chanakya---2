```python
import pandas as pd
```

```python
df = pd.read_csv('../../data/hopefully final master/hierarchy-pinky.csv')
```

```python
df = df.iloc[:, :11]
df
```

```python
res = pd.DataFrame(columns=df.columns.tolist())
for idx, row in df.iterrows():
    sizes = row['sizes'].split(',')
    for size in sizes:
        new_row = row.to_dict()
        new_row['sizes'] = size.strip()
        res = res.append(new_row, ignore_index=True)
    # break
res
```

```python
res.to_csv('../../data/hopefully final master/hierarchy-pinky-expanded.csv', index=False)
```
