```python
tigc_codes = set(tigc_master['EAN'].unique()) 
master_codes = set(master.ean.unique())

common_codes = tigc_codes & master_codes
```

```python
a = tigc_master[tigc_master['EAN'].isin(common_codes)][['EAN','Style Code','Product','Sub product','Brand','Fit']]
a.columns = ['ean','style_id','product','sub_product','brand','fit']
a.set_index('ean', inplace=True)
a.sort_index(inplace=True)
a
```

```python
b = master[master['ean'].isin(common_codes)][['ean','style_id','product','sub_product','brand','fit']]
b.ean = b.ean.astype(int)
b.set_index('ean', inplace=True)
b.sort_index(inplace=True)
b
```

```python
# product id
# fabric category?
# createdat, updatedat
new_styles = master[master.product_id.isnull()].style_id.unique()
```

```python
styles = {}
issues = {}
for style in new_styles:
    all_sizes = master[master.style_id==style].standard_size.values
    if len(all_sizes) != len(set(all_sizes)):
        print(all_sizes)
        issues[style] = all_sizes
    else:
        styles[style] = all_sizes 
```

```python
import uuid
# create a product id for each style and add it to master table
for style in new_styles:
    product_id = str(uuid.uuid4())
    master.loc[master.style_id==style, 'product_id'] = product_id
```
