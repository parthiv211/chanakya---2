```python
import json
import pandas as pd
```

```python
master = pd.read_csv('../../data/master.csv')
```

```python
fixed_cols = ['gender', 'vertical', 'fabric_category', 'brand', 'usage', 'brick','product', 'sub_product', 'look', 'target_audience', 'fit',
       'standard_size', 
        'colour_family', 'primary_colour', 'secondary_colour','tertiary_colour', 'season', 'exclusive',
        'garment_pattern','print_pattern_type', 'number_of_components', 'number_of_pockets','pocket_type', 'neck', 
        'collar', 'placket', 'length','sleeve_length','sleeve_type', 'hemline', 'waist_rise', 'closure', 'footwear_insole',
        'fabric_story', 'fabric_weave_pattern','wash_care','footwear_upper_material', 'footwear_sole_material',]
freetext_cols = ['style_id', 'barcode_size','tags','fabric_code','fabric_rate','fabric_composition', 'fabric_hsn','fabric_vendor','denim_cast','denim_wash', 'shoe_weight']
generated_cols = ['product_id','garment_waist',
       'inseam_length', 'to_fit_waist', 'across_shoulder', 'chest',
       'front_length', 'to_fit_foot_length', 'status',
       'created_at', 'created_by', 'updated_at', 'updated_by', 'deleted']
uploaded_cols = ['ean','myntra_id', 'ajio_id', 'ajio_ean', 'mrp', 'cost','first_grn_date', 'first_live_date', 'first_sold_date','images']
coltypes = {
    'fixed_cols': fixed_cols,
    'freetext_cols': freetext_cols,
    'generated_cols': generated_cols,
    'uploaded_cols': uploaded_cols,
}
print(coltypes)
# len(fixed_cols+generated_cols+freetext_cols+later_uploaded_cols)
```

```python
fixed_uniq = {}
for col in fixed_cols:
    fixed_uniq[col] = sorted([x for x in list(master[col].unique()) if x == x])
fixed_uniq
```

```python
with open('filters.json', 'w') as f:
    json.dump(fixed_uniq, f)
# fixed_uniq
```

```python
import numpy as np
hie = master.groupby(['gender', 'vertical', 'fabric_category', 'brand', 'usage', 'brick','product', 'sub_product', 'look', 'target_audience', 'fit',
       'standard_size']).mean()[['garment_waist',
       'inseam_length', 'to_fit_waist', 'across_shoulder', 'chest',
       'front_length', 'to_fit_foot_length']]
hie.replace(np.nan,'',regex=True, inplace=True)
hie.reset_index(inplace=True)
hie
```

```python
hie = hie.to_json(orient='records')
hie = json.loads(hie)
hie = {""}
with open('hie.json', 'w') as f:
    json.dump(hie, f)
```
