```python
def func(x):
    return ' '.join([y[0].upper() + y[1:].lower() for y in x.split()])
    x = set(x.split(','))
    if '' in x:
        x.remove('')
    return ', '.join(x)

styles["colour_family"] = styles["colour_family"].apply(func)
```
